```python
from PIL import Image
Image.open('onlline.png')
```

```python
img=Image.open('onlline.png')
# The file format of the source file.
print(img.format) # Output:JPEG

# The pixel format used by the image.
# Typical values are "I", "L", "RGB", or "ONYX."
print(img.mode) # Output: RGB

# Image size, in pixels
print(img.size) #Output: (1920, 1200)

print(img.palette) # Output: None
```
