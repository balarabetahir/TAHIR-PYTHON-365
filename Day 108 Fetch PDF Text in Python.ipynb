```python
import pdfplumber as pdftool
def Fetct_Text(file_path):
    # extract page by page
    with pdftool.open(file_path) as tool:
        for p_no, page in enumerate(tool.pages, 1):
            print('<--- page no', p_no, '--->')
            data = page.extract_text()
            print(data)
Fetct_Text("TA.pdf")
```
